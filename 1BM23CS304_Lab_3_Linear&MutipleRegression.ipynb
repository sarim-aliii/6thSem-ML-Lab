#### Linear Regression

```python
import pandas as pd
import numpy as np

df = pd.read_csv("canada_per_capita_income.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
X = df[['year']]
y = df['per capita income (US$)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
print(regression.coef_)
print(regression.intercept_)
```

```python
print(regression.predict([[2020]]))
```

Salary Dataset

```python
df = pd.read_csv("salary.csv")
```

```python
df.head()
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
df['YearsExperience'] = df['YearsExperience'].fillna(df['YearsExperience'].mean())
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
print("Slope (coefficient):", regression.coef_[0])
print("Intercept:", regression.intercept_)
print("Predicted salary", regression.predict([[12]]))
```

#### Multiple Regression

Hiring Dataset

```python
df = pd.read_csv("hiring.csv")

df['experience'] = df['experience'].replace({
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10
})
```

```python
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')
df['experience'] = df['experience'].fillna(df['experience'].median())
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
```

```python
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
salary1 = model.predict([[2, 9, 6]])
salary2 = model.predict([[12, 10, 10]])

print("Predicted salary for (2 yr exp, 9 test, 6 interview):", salary1[0])
print("Predicted salary for (12 yr exp, 10 test, 10 interview):", salary2[0])
```

Company Dataset

```python
df = pd.read_csv("1000_Companies.csv")
```

```python
df = pd.get_dummies(df, columns=['State'], drop_first=True)
```

```python
X = df.drop('Profit', axis=1)
y = df['Profit']

regression = LinearRegression()
regression.fit(X, y)

print("Coefficients:", regression.coef_)
print("Intercept:", regression.intercept_)
```

```python
prediction = regression.predict([[91694.48, 515841.3, 11931.24, 1, 0]])

print("Predicted Profit:", prediction[0])
```
